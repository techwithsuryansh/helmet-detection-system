```python
import os
```

```python
dataset_path="dataset"
for root, dirs, files in os.walk(dataset_path):
    for f in files:
        if f.endswith(".yaml") or f.endswith(".txt") and "class" in f.lower():
            full = os.path.join(root, f)
            print(f"\nFound: {full}")
            with open(full) as file:
                print(file.read())
```

```python

label_path = "dataset/train/labels"

class_counts = {}

for f in os.listdir(label_path):
    if f.endswith(".txt"):
        with open(os.path.join(label_path, f)) as file:
            for line in file.readlines():
                cls = line.strip().split()[0]
                class_counts[cls] = class_counts.get(cls, 0) + 1

print("Class ID : Count")
for cls, count in sorted(class_counts.items()):
    print(f"  Class {cls} : {count} boxes")
    
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data="dataset/data.yaml",
    epochs=40,
    imgsz=832,
    batch=8,
    workers=2,          
    close_mosaic=10,  
    name="helmet_detection",
    cos_lr=True,
    device=0
)

print("Training complete!")
print(f"Best model saved at: runs/detect/helmet_model/weights/best.pt")
```
